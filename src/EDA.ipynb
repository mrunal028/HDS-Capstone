```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("heart_attack_dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# Basic information
print("Shape of dataset:", df.shape)
```

```python
print("\nColumn data types:\n", df.dtypes)
```

```python
# Basic information
print("\nMissing values:\n", df.isnull().sum())
```

```python
df.describe()
```

```python
# Categorical column summary
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    print(f"\nValue counts for {col}:\n", df[col].value_counts())
```

```python
# Create a histogram to visualize the distribution of the 'Age' column
# - bins=30: number of bins for the histogram
# - kde=True: includes a Kernel Density Estimate curve for smooth distribution
# - color='darkgreen': sets the color of the histogram bars and KDE line
sns.histplot(df["Age"], bins=30, kde=True, color="darkgreen")

# Set the title of the plot
plt.title("Distribution of Ages in Dataset")

# Label the x-axis
plt.xlabel("Age (in years)")

# Label the y-axis
plt.ylabel("Number of Individuals")

# Display the plot
plt.show()
```

```python

# Create a count plot to visualize the frequency of each gender category
# - x="Gender": specifies the categorical variable to be plotted on the x-axis
# - data=df: specifies the DataFrame containing the data
# - palette='Set2': applies a different color palette for visual distinction
sns.countplot(x="Gender", data=df, palette="Set2")

# Set the title of the plot
plt.title("Count of Individuals by Gender")

# Label the x-axis
plt.xlabel("Gender")

# Label the y-axis
plt.ylabel("Number of Individuals")

# Display the plot
plt.show()
```
